## Storytelling With Data

**Time is the limit**  
Comunicare efficacemente con i dati richiede tempo: per comprendere il contesto, analizzare il pubblico, costruire una storia chiara e iterare sulle visualizzazioni. L’iterazione è fondamentale per affinare il messaggio e migliorare l’impatto visivo. L’unico modo per eccellere in questo processo è praticare costantemente.

**Great Visualization step by step**
1. **Comprendi il contesto**. Chiarisci chi è il tuo pubblico, cosa deve sapere o fare, come comunicherai con loro e quali dati supportano il tuo messaggio. Usa strumenti come la **3-minute story**, la **Big Idea** e lo **storyboarding** per strutturare la tua narrazione e definire il flusso dei contenuti.

1. **Scegli una visualizzazione appropriata**. Per evidenziare un singolo valore, il testo semplice è l'opzione migliore. I grafici a linee funzionano bene per dati continui, mentre i grafici a barre sono ideali per dati categoriali (assicurandosi di avere una baseline a zero). La relazione che vuoi mostrare deve guidare la scelta del grafico. Evita grafici a torta, donut, 3D e doppi assi y, che complicano l’interpretazione visiva.

1. **Elimina il disordine**. Identifica e rimuovi gli elementi che non aggiungono valore informativo. Sfrutta i **principi della Gestalt** per capire come le persone percepiscono le informazioni e individuare ciò che può essere eliminato. Usa il contrasto in modo strategico, allinea gli elementi e mantieni spazi bianchi per rendere la visualizzazione più leggibile e comprensibile.

1. **Focalizza l'attenzione dove serve**. Utilizza **attributi preattentivi** come colore, dimensione e posizione per evidenziare gli elementi chiave. Questi strumenti aiutano a guidare lo sguardo del pubblico e migliorare la comprensione del messaggio. Valuta l’efficacia della tua visualizzazione con il test: dove cade per prima l’attenzione?

1. **Pensa come un designer**. Offri al pubblico segnali visivi per interpretare correttamente il messaggio: evidenzia le informazioni più importanti, rimuovi le distrazioni e crea una gerarchia visiva chiara. Mantieni il **design semplice e accessibile**, utilizzando etichette testuali per spiegare concetti complessi. ***Un’estetica curata aumenta la tolleranza agli eventuali limiti del design e facilita l’accettazione della visualizzazione***.

1. **Racconta una storia**. Struttura il racconto con un **inizio** (contesto), uno **sviluppo** (snodi narrativi, tensione) e una **conclusione** (call to action). Usa il conflitto per mantenere alta l’attenzione. **Scegli con cura l’ordine e il modo in cui narrare** (cronologico o anticipando la fine, orale o scritto). La ripetizione aiuta a rendere memorabile il messaggio. Strumenti come la logica verticale e orizzontale, il reverse storyboarding e la ricerca di nuove prospettive possono migliorare la chiarezza della comunicazione.

### The Importance of the Context

**Obiettivo del documento e paradosso del "Slideument"**  
Uno degli errori più comuni nella comunicazione dei dati è confondere l’**exploratory analysis** (presentare tutti i dati grezzi) con l’**explanatory analysis** (selezionare e sintetizzare le informazioni più rilevanti per il pubblico). Questo errore nasce dalla tentazione di mostrare tutto il lavoro svolto, ma è essenziale resistere e concentrarsi su ciò che conta per il pubblico.

Un’altra sfida è il paradosso del **"Slideument"**, ovvero il tentativo di creare un unico prodotto che serva sia per una presentazione dal vivo (che richiederebbe slide essenziali e sintetiche) sia per un documento da consultare autonomamente (che richiederebbe maggiori dettagli). Spesso, per ragioni di tempo, si tende a un compromesso che rischia di non soddisfare pienamente nessuna delle due esigenze.

**Tips & Tricks: 3-minute story & Big Idea**  
Per rendere la comunicazione più efficace, è utile applicare la tecnica del **"3-minute story"**: se avessi solo tre minuti per trasmettere il tuo messaggio, cosa diresti? Questo aiuta a distillare l’essenza della comunicazione. Ancora più sintetica è la **"Big Idea"**, che riduce il messaggio chiave a una singola frase.

Un metodo efficace per costruire un messaggio chiaro è ragionare in termini di **Who, What, How** ma, soprattutto, prima di buttarsi sulle slide, è fondamentale partire dallo **storyboarding**: ovvero creare una struttura visiva della comunicazione. Questo aiuta a organizzare le informazioni in modo logico e coerente, evitando di disperdersi nei dettagli.

### Clutter is your enemy

**Cognitive Load: la tassa dell'attenzione**

Il cognitive load in una presentazione si riferisce alla quantità di informazioni che il pubblico deve elaborare per comprendere il messaggio. Un carico cognitivo eccessivo può ostacolare l’attenzione e la memorizzazione, rendendo la comunicazione meno efficace.

**Tips & Tricks: Gestalt Principles of Visual Perception**  

Le Gestalt Principles of Visual Perception descrivono come il nostro cervello organizza e interpreta le informazioni visive, rendendole fondamentali per progettare presentazioni efficaci.
- **Proximity (Prossimità)**: elementi vicini tra loro vengono percepiti come correlati o appartenenti allo stesso gruppo. Ad esempio, raggruppare dati visivamente aiuta a creare connessioni logiche.

- **Similarity (Somiglianza)**: oggetti con caratteristiche simili (colore, forma, dimensione) vengono percepiti come appartenenti alla stessa categoria. Usare la stessa tonalità o stile visivo rafforza il legame tra elementi.

- **Enclosure (Inclusione)**: elementi racchiusi da un bordo, colore di sfondo o un’area distinta vengono percepiti come un gruppo. Questo aiuta a evidenziare concetti correlati senza bisogno di spiegazioni aggiuntive.

- **Closure (Chiusura)**: la mente tende a completare forme o schemi incompleti per creare un'immagine coerente. È utile nelle visualizzazioni minimaliste, dove il cervello riempie i gap.

- **Continuity (Continuità)**: elementi allineati lungo una linea o curva vengono percepiti come connessi e parte di un flusso logico. Questo principio è utile nei grafici e nei percorsi visivi.

- **Connection (Connessione)**: elementi uniti da linee o collegamenti visivi sono interpretati come appartenenti allo stesso gruppo, anche più della prossimità o della somiglianza. Esempi sono le linee nei diagrammi o nelle mappe concettuali.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Impostare lo stile di Seaborn
sns.set(style="whitegrid")


# Generare dati per 3 gruppi separati
np.random.seed(42)
group_1 = np.random.normal(loc=[2, 2], scale=0.3, size=(20, 2))  # Gruppo in basso a sinistra
group_2 = np.random.normal(loc=[6, 6], scale=0.3, size=(20, 2))  # Gruppo in alto a destra
group_3 = np.random.normal(loc=[10, 2], scale=0.3, size=(20, 2)) # Gruppo in basso a destra

# Unire i dati
data = np.vstack([group_1, group_2, group_3])
df = pd.DataFrame(data, columns=["x", "y"])
df["group"] = ["Gruppo 1"]*20 + ["Gruppo 2"]*20 + ["Gruppo 3"]*20  # Etichette

# Creare il plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="x", y="y", hue="group", palette=["#00256d", "#1e6bff", "#80acff"], s=100)

# Rimuovere legenda per enfatizzare la prossimità visiva
plt.legend().remove()

# Titolo del grafico
plt.title("Principi della Prossimità & Somiglianza - Gestalt", fontsize=14)
plt.show()
```

```python
from matplotlib.patches import Ellipse

# Creare il plot
fig, ax = plt.subplots(figsize=(8,6))
sns.scatterplot(data=df, x="x", y="y", hue="group", palette=["#00256d", "#1e6bff", "#80acff"], s=100, ax=ax)


# Aggiungere aree di inclusione (Ellissi trasparenti attorno ai gruppi)
ellipses = [
    Ellipse(xy=(2, 2), width=2.5, height=2.5, edgecolor="#00256d", facecolor="#00256d", alpha=0.2, lw=2),
    Ellipse(xy=(6, 6), width=2.5, height=2.5, edgecolor="#1e6bff", facecolor="#1e6bff", alpha=0.2, lw=2),
    Ellipse(xy=(10, 2), width=2.5, height=2.5, edgecolor="#80acff", facecolor="#80acff", alpha=0.2, lw=2)
]

# Aggiungere gli ellissi al grafico
for ellipse in ellipses:
    ax.add_patch(ellipse)

# Rimuovere legenda per enfatizzare la percezione visiva
plt.legend().remove()

# Titolo del grafico
plt.title("Prossimità, Somiglianza & Inclusione - Gestalt", fontsize=14)
plt.show()
```

```python
# Generare punti per una forma circolare incompleta
theta = np.linspace(0, 2*np.pi, 30)  # Angoli per creare il cerchio
x = np.cos(theta) + 5  # Sposta il cerchio al centro (5,5)
y = np.sin(theta) + 5

# Rimuovere alcuni punti per creare un effetto di "chiusura incompleta"
mask = (theta < 1) | (theta > 5)  # Mantiene solo parte del cerchio
x_partial, y_partial = x[mask], y[mask]

# Creare il plot
plt.figure(figsize=(6,6))
plt.scatter(x_partial, y_partial, color="#00256d", s=100)

# Aggiungere titolo
plt.title("Principio della Chiusura & Continuità - Gestalt", fontsize=14)

# Rimuovere assi per enfatizzare la forma
plt.xticks([])
plt.yticks([])
plt.box(False)

plt.show()
```

```python
# Creare il plot
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(x_partial, y_partial, color="#00256d", s=100, label="Chiusura")

# Aggiungere connessioni con linee tra alcuni punti per rafforzare il raggruppamento
for i in range(len(x_partial) - 1):
    ax.plot([x_partial[i], x_partial[i+1]], [y_partial[i], y_partial[i+1]], color="#80acff", linewidth=2)

# Aggiungere titolo
plt.title("Chiusura, Continuità & Connessione - Gestalt", fontsize=14)

# Rimuovere assi per enfatizzare la percezione visiva
plt.xticks([])
plt.yticks([])
plt.box(False)

# Mostrare il grafico
plt.show()
```

### Focus your audience attention

**Preattentive attribute**

Un preattentive attribute è una **caratteristica visiva che il nostro cervello processa istantaneamente e inconsciamente**, prima ancora di focalizzare l’attenzione. Questi attributi, come colore, forma, dimensione, orientamento e movimento, permettono di evidenziare informazioni chiave e guidare rapidamente l’attenzione dell’osservatore in una visualizzazione. Sfruttarli in grafici e presentazioni aiuta a rendere i dati più leggibili e intuitivi, riducendo il carico cognitivo. Oltre alla rapidità di interpretazione possono comunicare la struttura gerarchica dell'informazione.   
Operativamente un consiglio utile potrebbe essere quello di partire mettendo tutto in background e poi elevare gli elementi più rilevanti grazie a questi elementi. Ricorda infine che l'audience tende a partire dall'angolo in alto a sinistra e poi procede a zig zag verso l'angolo in basso a destra attraverso la diagonale.

```python
import numpy as np

# Creare una nuova colonna "group2" inizialmente uguale a "Gruppo 1"
df["group2_baseline"] = "Gruppo 1"
df["group2"] = "Gruppo 1"

# Selezionare casualmente 6 indici da modificare
random_indices = np.random.choice(df.index, size=6, replace=False)

# Assegnare "Preattentive attributes" a questi indici
df.loc[random_indices, "group2"] = "Preattentive attributes"
# Creare il plot
fig, ax = plt.subplots(figsize=(8,6))
sns.scatterplot(data=df, x="x", y="y", hue="group2_baseline", palette=["#cecece"], s=100, ax=ax)


# Aggiungere aree di inclusione (Ellissi trasparenti attorno ai gruppi)
ellipses = [
    Ellipse(xy=(2, 2), width=2.5, height=2.5, edgecolor="#cecece", facecolor="#cecece", alpha=0.2, lw=2),
    Ellipse(xy=(6, 6), width=2.5, height=2.5, edgecolor="#cecece", facecolor="#cecece", alpha=0.2, lw=2),
    Ellipse(xy=(10, 2), width=2.5, height=2.5, edgecolor="#cecece", facecolor="#cecece", alpha=0.2, lw=2)
]

# Aggiungere gli ellissi al grafico
for ellipse in ellipses:
    ax.add_patch(ellipse)

# Rimuovere legenda per enfatizzare la percezione visiva
plt.legend().remove()

# Disabilita la griglia
plt.grid(False)


# Titolo del grafico
plt.title("Preattentive Attributes", fontsize=14)
plt.show()
```

### Think like a designeer

**Form follows Function**

Adottare un approccio da designer nella visualizzazione dei dati migliora l'efficacia della comunicazione, rendendo le informazioni più intuitive e coinvolgenti. Studi dimostrano che design estetici e ben curati non solo facilitano l’uso, ma aumentano l'accettazione e la tolleranza verso eventuali imperfezioni, favorendo la creatività e il problem solving. Il principio "form follows function" si applica anche ai dati: prima si definisce l’obiettivo della comunicazione (funzione), poi si progetta la visualizzazione (forma) per massimizzare la comprensione. Integrare concetti di design, come la gerarchia visiva, la rimozione di elementi superflui e l’uso di affordances visive, aiuta il pubblico a navigare le informazioni con maggiore facilità. Inoltre, un aspetto visivamente curato può aumentare la pazienza e l'engagement dell’audience, migliorando l’efficacia complessiva della comunicazione.

### Lessons in storytelling

**How to structure a story**  
La struttura narrativa classica si basa su **setup, conflitto e risoluzione**. Il primo atto introduce il protagonista, il suo mondo e le sue relazioni. Un incidente scatenante porta a un cambiamento significativo, culminando in un primo punto di svolta, che trasforma la storia e pone una domanda centrale che troverà risposta nel climax.

Per costruire una narrazione efficace nella visualizzazione dei dati, è essenziale strutturare la comunicazione con un inizio (contesto), uno sviluppo (analisi) e una conclusione (call to action), utilizzando conflitto e tensione per mantenere alta l’attenzione del pubblico ma il vero protagonista della storia deve essere il pubblico, rendendo i dati rilevanti e significativi per chi ascolta, trasformandoli in un punto chiave della narrazione anziché una semplice esposizione di numeri. Per garantire un flusso logico, si può partire dalla scrittura dei titoli delle sezioni (headline-first approach), riordinandoli fino a creare una sequenza chiara e coerente, come si farebbe in uno storyboard.

**Simplify and sounds like yourself**  
Scrivere in modo efficace significa essere semplici, eliminare il superfluo e mantenere un tono autentico. Ogni frase deve illuminare il messaggio, altrimenti va tagliata. Bisogna esprimersi con chiarezza, rispettando le regole del linguaggio per garantire comprensibilità. Infine, occorre avere empatia per il lettore, semplificando, ripetendo e chiarendo il messaggio per facilitare la comprensione.
